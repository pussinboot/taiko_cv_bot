## messup detection

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from skimage import io
import cv2
```

```python
messup_img = io.imread('../masks/messup.png')
plt.imshow(messup_img)
plt.show()
```

```python
plt.imshow(messup_img[35:67,101:153,:]) # now we know the offsets from top left corner 
plt.show()
```

```python
messup = io.imread('../masks/messup.png')[35:67,101:153,:3]
r1, g1, b1 = 0, 255, 0
r, g, b = messup[:,:,0], messup[:,:,1], messup[:,:,2]
mask = (r == r1) & (g == g1) & (b == b1)
messup_mask = 1-mask
messup[mask] = [-1,-1,-1]
plt.subplot(1,2,1)
plt.imshow(messup)
plt.subplot(1,2,2)
plt.imshow(messup_mask)
plt.show()
```

```python
test_frame_hit = io.imread('../extracted/out-15.png') # a different frame from the one where i cut out the messup img
test_frame_miss = io.imread('../extracted/out-60.png') # a frame without messup
plt.subplot(1,2,1)
plt.imshow(test_frame_hit)
plt.subplot(1,2,2)
plt.imshow(test_frame_miss)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.imshow(test_frame_hit[35:67,101:153,:])
plt.subplot(1,2,2)
plt.imshow(test_frame_miss[35:67,101:153,:])
plt.show()
```

```python
test_hit = test_frame_hit[35:67,101:153,:][messup_mask]
test_miss = test_frame_miss[35:67,101:153,:][messup_mask]
print(np.mean(cv2.absdiff(test_frame_miss[35:67,101:153,:],messup)))#,np.mean(cv2.absdiff(test_miss,messup[messup_mask])))
```

```python
plt.subplot(2,1,1)
plt.imshow(test_frame_hit[35:67,101:153,:]-messup)
plt.subplot(2,1,2)
plt.imshow(test_frame_miss[35:67,101:153,:]-messup)
plt.show()
```

```python
another_test_frame = io.imread('../extracted/out-7.png')
plt.imshow(another_test_frame[35:67,101:153,:]-messup)
plt.show()
```

```python
from itertools import chain
test_samples = chain(range(1,51),range(199,228),range(1000,1062))
means_with_mask = []
means_without_mask = []
for i in test_samples:
    test_frame = io.imread('../extracted/out-{}.png'.format(i))
    means_with_mask += [np.mean(cv2.absdiff(test_frame[35:67,101:153,:][messup_mask],messup[messup_mask]))]
    means_without_mask += [np.mean(cv2.absdiff(test_frame[35:67,101:153,:],messup))]
```

```python
plt.plot(means_with_mask,'b-',label='w/ mask')
plt.plot(means_without_mask,'r-',label='w/o mask')
plt.legend(loc='center')
plt.show()
```

```python
# compare to frames without messup : )
miss_means_with_mask = []
miss_means_without_mask = []
for i in range(50,199):
    test_frame = io.imread('../extracted/out-{}.png'.format(i))
    miss_means_with_mask += [np.mean(cv2.absdiff(test_frame[35:67,101:153,:][messup_mask],messup[messup_mask]))]
    miss_means_without_mask += [np.mean(cv2.absdiff(test_frame[35:67,101:153,:],messup))]
```

```python
plt.plot(miss_means_with_mask,'b-',label='w/ mask')
plt.plot(miss_means_without_mask,'r-',label='w/o mask')
plt.legend(loc='lower center')
plt.show()
```

```python
# looks like without mask works better :L
plt.plot(means_without_mask,'b-',label='hit')
plt.plot(miss_means_without_mask[:len(means_without_mask)],'r-',label='miss')
plt.legend(loc='center')
plt.text(5,90,'bubble fading out along here')
plt.show()
```

```python
print(max(means_without_mask), np.mean(miss_means_without_mask))
```
