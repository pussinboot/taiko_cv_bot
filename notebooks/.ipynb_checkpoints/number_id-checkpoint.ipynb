```python
# preprocess img to be nice
r1, g1, b1 = 0, 255, 0
r, g, b, a = zeroimg[:,:,0], zeroimg[:,:,1], zeroimg[:,:,2], zeroimg[:,:,3]
mask = (r == r1) & (g == g1) & (b == b1)
zeroimg[mask] = [0,0,0,255]
```

```python
plt.imshow(zeroimg)
```

```python
for i in range(5):
    print(i,np.mean(cv2.absdiff(testimg[:,i*12:(i+1)*12,:],zeroimg)))
# numbers without fixing up image
```

```python
for i in range(5):
    print(i,np.mean(cv2.absdiff(testimg[:,i*12:(i+1)*12,:],zeroimg)))
# numbers WITH fixing up image
# higher variance i guess lol
```

```python
# now let's build this up for all the numbers
no_imgs = [None]*10
for i in range(10):
    filename = '../masks/{}.png'.format(i)
    no_imgs[i] = io.imread(filename)
```

```python
for i in range(len(no_imgs)):
    print(i,np.mean(cv2.absdiff(testimg[:,12:24,:],no_imgs[i])))
# without cleaning the images
```

```python
new_imgs = [None]*10
r1, g1, b1 = 0, 255, 0
for i in range(len(no_imgs)):
    new_imgs[i] = no_imgs[i]
    r, g, b, a = no_imgs[i][:,:,0], no_imgs[i][:,:,1], no_imgs[i][:,:,2], no_imgs[i][:,:,3]
    mask = (r == r1) & (g == g1) & (b == b1)
    new_imgs[i][mask] = [0,0,0,255]
```

```python
for i in range(len(no_imgs)):
    print(i,np.mean(cv2.absdiff(testimg[:,12:24,:],new_imgs[i])))
# with cleaned images
```

```python
# looks like cleaned images win sry
# so now let's figure out where to look in a real game :]
for i in range(len(no_imgs)):
    no_imgs[i] = no_imgs[i][:,:,:3]
    new_imgs[i] = new_imgs[i][:,:,:3] # need this to compare with real screenshot ;-;
```

```python
# real_screenshot_fname = """../masks/1413 - Taiko no Tatsujin DS - Touch de Dokodon! (v01) (J)(Independent)__429.png"""
# let's try some more screenshots
# real_screenshot_fname = """../masks/1413 - Taiko no Tatsujin DS - Touch de Dokodon! (v01) (J)(Independent)__28055.png"""
# real_screenshot_fname = """../masks/1413 - Taiko no Tatsujin DS - Touch de Dokodon! (v01) (J)(Independent)__28192.png"""
real_screenshot_fname = """../masks/1413 - Taiko no Tatsujin DS - Touch de Dokodon! (v01) (J)(Independent)__28330.png"""

screenshot = io.imread(real_screenshot_fname)
```

```python
testsegment = screenshot[52:68,-94:-10,:]
```

```python
# go from right to left, find if a number is below the threshold, if none work then exit out
# looks like nos without processing work better at this point hmm
res = []
for i in reversed(range(7)):
    arr = [np.mean(cv2.absdiff(testsegment[:,12*i:12*(i+1),:],no_imgs[n])) \
         for n in range(len(new_imgs))]
    if min(arr) > 75: break
    res += [np.argmin(arr)]
    #print(min(arr))
res.reverse()
int(''.join(map(str, res)))
```
