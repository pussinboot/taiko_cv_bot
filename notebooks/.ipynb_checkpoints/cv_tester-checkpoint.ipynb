```python
# %load id_number.py
import numpy as np
from skimage import io
import cv2

class NumberRecog:
	def __init__(self,path_to_files='../masks'):
		self.no_imgs = [None]*10
		self.last_score = 0
		for i in range(10):
			filename = path_to_files+'/{}.png'.format(i)
			self.no_imgs[i] = io.imread(filename)[:,:,:3]
			r1, g1, b1 = 0, 255, 0
			r, g, b, a = self.no_imgs[i][:,:,0], self.no_imgs[i][:,:,1], self.no_imgs[i][:,:,2]
			mask = (r == 0) & (g == 255) & (b == 0)
			self.no_imgs[i][mask] = [0,0,0]

	def clean_img(self,img_to_clean,thresh=240):
		r, g, b = img_to_clean[:,:,0], img_to_clean[:,:,1], img_to_clean[:,:,2]
		mask = (r < thresh) & (g < thresh) & (b < thresh)
		img_to_clean[mask] = [0,0,0]
		return img_to_clean

	def test_img(self,img_to_test):
		res = []
		img_to_test = self.clean_img(img_to_test)
		for i in reversed(range(7)):
		    arr = [np.mean(cv2.absdiff(img_to_test[:,12*i:12*(i+1),:],self.no_imgs[n])) \
		         for n in range(len(self.no_imgs))]
		    if min(arr) > 50: break
		    res += [np.argmin(arr)]
		    #print(min(arr))
		res.reverse()
		torstr = ''.join(map(str, res))
		tor = int(torstr)
# 		if abs(len(torstr) - len(str(self.last_score))) > 2:
# 			if self.last_score > 0:
# 				tor = self.last_score
# 		self.last_score = tor
		return tor

	def test(self,img_to_test):
		try:
			return self.test_img(img_to_test)
		except:
			return -1

if __name__ == '__main__':
	testng = NumberRecog()
```

```python
NR = NumberRecog('../masks')
cap = cv2.VideoCapture(1)
cap.set(3,256)
cap.set(4,384)
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    res = str(NR.test(frame[52:68,-94:-10,:]))
    cv2.putText(frame,res,(0,220), cv2.FONT_HERSHEY_DUPLEX, 1,(0,255,0),2)
    # Display the resulting frame
    cv2.imshow('frame',frame)
   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    elif cv2.waitKey(1) & 0xFF == ord('o'):
        print(res)

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
