```python
good_crop = io.imread('../masks/good.png')
okay_crop = io.imread('../masks/okay.png')
bad_crop = io.imread('../masks/bad.png')
```

```python
	def clean_img(img_to_clean):
		rgb = [0,255,0]
		mask = np.where(np.all(img_to_clean == rgb,axis=-1))
		img_to_clean[mask] = [-1,-1,-1]
		return img_to_clean
```

```python
good_mask = clean_img(good_crop[:,:,:3])
okay_mask = clean_img(okay_crop[:,:,:3])
bad_mask = clean_img(bad_crop[:,:,:3])
plt.subplot(1,3,1)
plt.imshow(good_mask)
plt.subplot(1,3,2)
plt.imshow(okay_mask)
plt.subplot(1,3,3)
plt.imshow(bad_mask)
```

```python
# the offset of these are
# [50:75,15:45,:3]
# of a frame
plt.imshow(good[50:75,15:45,:3]-good_mask) # : )
```

```python
#let's test
def test_img(img_to_test,compare):
   return np.mean(cv2.absdiff(img_to_test[50:75,15:45,:3],compare))
```

```python
masks =  [good_mask, okay_mask, bad_mask]   
frames = [good, okay, bad]
names = ['good','okay','bad']
for i in range(3):
    res = [test_img(frames[i],mask) for mask in masks]
    print('{0}, {1}'.format(names[i],res))
```

```python
test_good_okay_bad = np.zeros([3251,3])
```

```python
masks =  [good_mask, okay_mask, bad_mask]   

for i in range(1,3252):
    filename = '../extracted/out-{}.png'.format(i)
    testframe = io.imread(filename)
    test_good_okay_bad[i-1,:] = [test_img(testframe,mask) for mask in masks]
```

```python
plt.figure(figsize=(14,5))
plt.plot(test_good_okay_bad[:,0],'g-',label='good')
plt.plot(test_good_okay_bad[:,1],'b-',label='okay')
plt.plot(test_good_okay_bad[:,2],'r-',label='bad')
ax = plt.axis([0, 3251, 70, 150])
plt.plot((0,3251),(100,100),'k--',label='thresh')
plt.legend(loc='upper right')
plt.show()
```

in order to do proper detection of these signals, can't just say less than a certain threshold means it is detected since would get multiple detections per actual event

so, instead have to have a reset for each signal so that a good hit can "fire" 

```python
a = 50
b = 200
plt.plot(test_good_okay_bad[a:b,0],'g-',label='good')
plt.plot(test_good_okay_bad[a:b,1],'b-',label='okay')
plt.plot(test_good_okay_bad[a:b,2],'r-',label='bad')
plt.show()
```
